```python
# contoh parametric probability density estimation
# library yang digunakan 

from matplotlib import pyplot
from numpy.random import normal
from numpy import mean
from numpy import std
from scipy.stats import norm

# membangkitkan sample 
sample = normal(loc=50, scale=5, size=1000)

# menghitung parameters
sample_mean = mean(sample)
sample_std = std(sample)
print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))

# mendefinisikan distribusi  
dist = norm(sample_mean, sample_std)
values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values]

# plotting histogram and pdf
pyplot.hist(sample, bins=10, density=True)
pyplot.plot(values, probabilities)
pyplot.show()
```

```python
# Library yang diguanakan 
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
from numpy import asarray
from numpy import exp
from sklearn.neighbors import KernelDensity
# contoh kondisi bimodal data
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
# membangkitkan sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = hstack((sample1, sample2))
# plot histogram
pyplot.hist(sample, bins=50)
pyplot.show()

```

```python
#Untuk melakukan fitting terhadap sample tersebut, kita gunakan fungsi kernal density estimation 

model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)

# sample probabilities for a range of outcomes
values = asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)

# plot the histogram and pdf
pyplot.hist(sample, bins=50, density=True)
pyplot.plot(values[:], probabilities)
pyplot.show()
```

Pada estimasi kepekatan parametrik, cari distribusi lain selain distribusi normal dan lakukan visualiasi

```python
from scipy.stats import logistic
import matplotlib.pyplot as plt
import numpy as np

# parameter logistic distribution
loc = 50  # location (mean)
scale = 5  # scale (standard deviation)

# buat sample logistic distribution
sample = logistic.rvs(loc=loc, scale=scale, size=1000)

# Plot histogram dari sample
plt.hist(sample, bins=30, density=True)

# Plot PDF logistic distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = logistic.pdf(x, loc=loc, scale=scale)
plt.plot(x, p)
plt.show()
```

```python
plt.scatter(range(len(sample)), sample)

# legend
plt.title('Data Sample')
plt.xlabel('index')
plt.ylabel('Values')

# Show the plot
plt.show()
```

Buat histogram untuk contoh pembangkitan data dengan jumlah peak = 3 (multimodal) kemudian gunakan fungsi kernel density estimation dan lakukan visualisasi

```python
# Library yang diguanakan 
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
from numpy import asarray
from numpy import exp
from sklearn.neighbors import KernelDensity
# membangkitkan sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample3 = normal(loc=60, scale=5, size=500)
sample = hstack((sample1, sample2, sample3))
# plot histogram
pyplot.hist(sample, bins=50)
pyplot.show()
#Untuk melakukan fitting terhadap sample tersebut, kita gunakan fungsi kernal density estimation 
model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)
# sample probabilities for a range of outcomes
values = asarray([value for value in range(1, 80)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)
# plot the histogram and pdf
pyplot.hist(sample, bins=50, density=True)
pyplot.plot(values[:], probabilities)
pyplot.show()
```
